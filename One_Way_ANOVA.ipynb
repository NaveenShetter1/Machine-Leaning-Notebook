```python
import numpy as np
from scipy import stats

# Sample data: three different groups
group1 = [23, 21, 18, 25, 30]
group2 = [33, 30, 35, 40, 38]
group3 = [43, 42, 41, 45, 44]

# Perform One-Way ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

print("F-statistic:", f_statistic)
print("P-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: At least one group mean is different.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between group means.")
```

```python
import matplotlib.pyplot as plt

# Combine the data into a list of groups
data = [group1, group2, group3]

# Create a boxplot to visualize the group differences
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('One-Way ANOVA')
plt.ylabel('Value')
plt.show()
```
